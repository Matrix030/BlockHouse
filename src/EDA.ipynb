```python
first_row = df.iloc[0]
print("\nAPPLE (AAPL) Order Book Snapshot:")
print(f"Timestamp: {first_row['ts_recv']}")
print("\n=== ORDER BOOK (10 levels deep) ===")
print("BIDS (Buyers)           |  ASKS (Sellers)")
print("Price    Size  Count    |  Price    Size  Count")
print("------------------------|-----------------------")

for i in range(10):
   level = f"{i:02d}"
   bid_px = first_row[f'bid_px_{level}']
   bid_sz = first_row[f'bid_sz_{level}']
   bid_ct = first_row[f'bid_ct_{level}']
   ask_px = first_row[f'ask_px_{level}']
   ask_sz = first_row[f'ask_sz_{level}']
   ask_ct = first_row[f'ask_ct_{level}']
   
   print(f"{bid_px:<8} {bid_sz:>4} {bid_ct:>3}    |  {ask_px:<8} {ask_sz:>4} {ask_ct:>3}")

spread = first_row['ask_px_00'] - first_row['bid_px_00']
mid_price = (first_row['ask_px_00'] + first_row['bid_px_00']) / 2

print(f"\nSpread (Ask - Bid): {spread:.2f}")
print(f"Mid Price: {mid_price:.2f}")
```

```python
# Look at time progression
print("\nFirst 10 market events:")
print("Row | Action | Side | Price | Best Bid | Best Ask")
print("----|--------|------|-------|----------|----------")

for i in range(10):
   row = df.iloc[i]
   print(f"{i:3d} | {row['action']:>6} | {row['side']:>4} | {row['price']:>6} | {row['bid_px_00']:>8} | {row['ask_px_00']:>8}")
```

```python
# Check unique values
print(f"\nUnique actions in data: {df['action'].unique()}")
print(f"Sides (B=Bid/Buy, A=Ask/Sell): {df['side'].unique()}")
print(f"Symbols: {df['symbol'].unique()}")
```

```python
# Basic statistics on prices
print(f"\nPrice range: ${df['price'].min():.2f} - ${df['price'].max():.2f}")
print(f"Best bid range: ${df['bid_px_00'].min():.2f} - ${df['bid_px_00'].max():.2f}")
print(f"Best ask range: ${df['ask_px_00'].min():.2f} - ${df['ask_px_00'].max():.2f}")
```

```python
# Count of each action type
print(f"\nAction counts:")
print(df['action'].value_counts())
```

```python
print(f"\nSide counts:")
print(df['side'].value_counts())
```

```python
# Time span analysis
df['ts_recv'] = pd.to_datetime(df['ts_recv'])
time_span = df['ts_recv'].max() - df['ts_recv'].min()
print(f"\nData time span: {time_span}")
print(f"From: {df['ts_recv'].min()}")
print(f"To: {df['ts_recv'].max()}")
```
